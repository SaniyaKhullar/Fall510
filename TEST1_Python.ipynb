```python
#YAY SO FAR! :)


#coordinateValues = []
def polygarea(coords):
    """
    Please describe the function. 
    
    Please note that for xy coordinates in the positive ++ quadrant (quadrant 1), Green's Theorem returns Positive area ONLY if the vertices are numbered clockwise around the perimeter.  That is, sketching the polygon should be done in a clockwise way. 
    If it is done in a counter-clockwise ordering sketch, then the computed area will be negative. 
    
    For example:
    polygarea([(2, 2), (3,2), (2,1)]) yields a negative area: -1
    polygarea([(3, 2), (2,2), (2,1)]) yields a positive area: 1
    

    In both cases, the set of points are the same.  However the computed area is different based on the order. Please note that here, for this function, Saniya (based upon Professor Pilcher's formula) is using the absolute value, so that no matter what the area returned will ALWAYS be positive :).  No need to worry about clockwise or counterclockwise here...please order polygon sides as ye like, but in consecutive edge order:)
    Please note these characteristics of a polygon:
        Sides do NOT intersect. 
        Closed figure
        All edges are straight lines. 
        
    Please note that here Saniya has put the comments in terms of this general example (given by Professor Pilcher)
     of this square with these coordinates:  (0, 0), (2,0), (2,2), (0,2). 
     Please note that the coordinates should be entered in the order in which the polygon should be created.  All coordinates should be unique.  The first set of coordinates essentially becomes interpreted as the last set of coordinates as well since the figure needs to close. 
     Therefore, for these coordinates: (0, 0), (2,0), (2,2), (0,2), the program will go from: (0,0) to (2,0) to (2,2) to (0,2), and finally back to (0,0).
     
     Please note that when entering the coordinates in the polygarea function, please enter them in this form:
     polygarea([(x1,y1), (x2,y2), (x3,y3),...(xn,yn)]).
     That is for example the 2 by 2 square formed by these points: (0, 0), (2,0), (2,2), (0,2) will have an area of: (2)(2)= 4 units squared.
    Hence, the function: polygarea([(0, 0), (2,0), (2,2), (0,2)]) should yield a 4. 
    
    Please note that in the comments, oftentimes ye will see "print" commented out and this is since Saniya would test the code values for different parts.  It is helpful to still include these comments so that the reader can understand Saniya's thought processes. 
    Please note that Saniya humbly then tested the results obtained with (X,Y) polygon area calculators on the web to see how the results compared and those matched. 
    """
    CoordsList = [num for elem in coords for num in elem]   # flatten a list using a listcomp with two 'for'
    # please note the source of this code idea is: #https://docs.python.org/2/tutorial/datastructures.html
    # we thus get: [0, 0, 2, 0, 2, 2, 0, 2]
    #then, we please assign this to a vector: 


    # even index elements go to the x matrix.  Even though 0 is NOT even, the elements 2, 4, etc. are! 
    # odd index elements go to the y matrix. 

    LengthCoordsList = len(CoordsList)
    #print(LengthCoordsList) #8 here

    if LengthCoordsList == 4:
        print("Please enter 3 or more (x,y) coordinates!")

    NumOfCoordinatesN = (LengthCoordsList)/(2) #please note that this is the value for n
    #print(NumOfCoordinatesN)  #here, this would yield 4 since there are 4 coordinates! :)
    xlist = []
    ylist = []

    xlistForPosTerms = []  #please remove the last element here for the xlist
    xlistForNegTerms = []

    i = 0

    while i < LengthCoordsList:
        NewXvalue = CoordsList[i]
        xlist.append(NewXvalue)
        xlistForPosTerms.append(NewXvalue)
        xlistForNegTerms.append(NewXvalue)

        i = i + 2
    firstxValue = CoordsList[0]
    xlist.append(firstxValue)  #please note that we want to join the coordinates of the figure in the order they are given and so we must go back to square 1 (pun intended), which is the 1st point given.  We are closing the figure. 
    xlistForNegTerms.append(firstxValue) 

    #print(xlist)  #the 1st and last element will correspond to the same x,y pair...here we are closing the figure and the x component goes to the xlist.
    #print(xlistForPosTerms, "POS and misses last element")  
    #print(xlistForNegTerms, "NEG")



    j = 1  #here, the y values start at index 1 and go on to index 3, 5, ... etc. 
#the same concept applies to creating the vectors of y values. 
    ylistForPosTerms = []
    ylistForNegTerms = []
    while j < LengthCoordsList:
        NewYvalue = CoordsList[j]
        ylist.append(NewYvalue) 
        ylistForPosTerms.append(NewYvalue) 
        ylistForNegTerms.append(NewYvalue) 
        j = j + 2
    #adding on the last element, which should be the same as the 1st element. 
    firstyValue = CoordsList[1]
    ylist.append(firstyValue)
    ylistForPosTerms.append(firstyValue)

    #print(ylistForPosTerms, "POS")  
    #print(ylist)
    #print(ylistForNegTerms, "NEG and misses last element")


    #defensive programming:  just in case, please note
    if len(xlist) != len(ylist):
        print("Please note that there is a problem with the coordinate values you have entered! Please make sure you have specified both an x and y coordinate for each x,y pair.")
    
    
    
    
    #print(CoordsList, "CoordsList")
    #print(4 + 5)

    #to the xlist we will additionally append the 1st element at the end since Greens Theorem goes back to the start. 

    # Please note: An index of x 1 greater than y leads to negative term like:  -x2y1, -x3y2, -x4y3, -x5y4


    maxIndex = len(xlist) #here, please note that this is 5. 


    #print(maxIndex, "maxIndex")

    # Please note: An index of x 1 less than y leads to a positive term like:  x1y2, x2y3, x3y4, x4y5


    maxIndex = len(xlist) 
    #print(maxIndex, "maxIndex")
    
    #here, we please try to find the positive terms like x1y2, x2y3, x3y4, x4y5, etc. 


    # Green's Theorem involves summing these values to get some net number...then dividing that net number by 2 yields the Area



    #***Below, please note that we please try to find the positive terms like x1y2, x2y3, x3y4, x4y5, etc. 

    #please note that the pop function is of the form:  list.pop(index), where index is the position of the element in the list that we want to remove. 
    #the pop function pops off that given element with that index position in the original list.  It gets rid of that value.  
    indexLastElement = len(xlist) - 1  #please note that the indexing begins at 0.  Therefore, a list of 5 elements will go up to index 4.  Hence, we please subtract 1 from the # of elements (given by len(xlist)) to get the maximum index. 
    #print(indexLastElement)


    #xlist.pop(indexLastElement)
    #print(xlistForPosTerms)
    #print(xlistForPosTerms, "xlistposterms")

    #please remove the 1st element here for the ylist.  Please note that this 1st element always has the index of 0. 



    ylistForPosTerms.pop(0)  #please note this is removing the 1st term from the ylist. 
    #print(ylistForPosTerms, "ylistposterms")


    #now, please note that these are the 2 new lists: 

    #xlistForPosTerms = [0, 2, 2, 0] -->  values are x1, x2, x3, and x4 respectively (From the original xlist)
    # ylistForPosTerms = [0, 2, 2, 0]  --> values are y2, y3, y4, and y5 respectively (From the original ylist)

    #Now, when we multiply respective elements in xlistForPosTerms with the corresponding ones in ylistForPosTerms, we essentially will get: 
    # a list with:  [x1y2, x2y3, x3y4, x4y5,..., etc.] 

    xlistMultylistPosValues = []  # Please note that this creates an empty list that will ultimately have x1y2, x2y3, x3y4, x4y5, etc. 

    for k in range(0, len(xlistForPosTerms)):
        xlistMultylistPosValues.append(xlistForPosTerms[k] * ylistForPosTerms[k])

    #print(xlistMultylistPosValues) #[0, 4, 4, 0]

    #Then to get the sum of all the positive values, please note that we will just use the sum(list) function and that adds together ALL of the terms in the list. 

    sumPositivexyvalues = sum(xlistMultylistPosValues)
    #print(sumPositivexyvalues)

    #hence, sumPositivexyvalues = x1y2 + x2y3 + x3y4 + x4y5 + ... etc.




    #***Below, please note that we now please try to find the negative terms like x2y1, x3y2, x4y3, x5y4, etc. 

    # indexLastElement gives us the maximum index in the ylist (and xlist).  
    #Here, we want to please remove the 1st element from the xlist and the last element from the ylist. 


    #Please remove the 1st element from the xlist: 

    #print(xlistForNegTerms, "xlistnegtermsbefore")
    xlistForNegTerms.pop(0)

    #print(xlistForNegTerms, "xlistnegterms now")
    #xlistForNegTerms.append()


    #print(ylistForNegTerms, "ylistnegterms")

    xlistMultylistNegValues = []  #creates an empty list that we will use to please store the elements: x2y1, x3y2, x4y3, x5y4, etc.
    for g in range(0, len(xlistForNegTerms)):  #xlistForNegTerms has the terms x2, x3, x4, x5,...  while ylistForNegTerms has the terms y1, y2, y3, y4, etc...
        xlistMultylistNegValues.append(xlistForNegTerms[g] * ylistForNegTerms[g])

    #print(xlistMultylistNegValues, "xmulty neg terms")



    #Then to get the sum of all the negative values, please note that we will just use the sum(list) function and that adds together ALL of the terms in the list. 

    sumNegativexyvalues = sum(xlistMultylistNegValues)
    #print(sumNegativexyvalues)

    #hence, sumNegativexyvalues = x2y1 + x3y2 + x4y3 + x5y4 + ... etc.

    #Please note that since the sumNegativexyvalues should be negative, we multiply it by negative 1 to get or subtract it from the sumPositivexyvalues to get:

    SumWithinBracketsGreensTheorem = sumPositivexyvalues - sumNegativexyvalues
    #print(SumWithinBracketsGreensTheorem, "= sum within brackets using Green's Theorem")

    #Please note that manually, we found Green's area returned a negative value as well.  
    # We please convert this sum in brackets into a positive value since we please take the absolute value here. 
    
    
    if SumWithinBracketsGreensTheorem < 0:
        PositiveSumWithinBracketsGreensTheorem = -1 * SumWithinBracketsGreensTheorem
    else:
        PositiveSumWithinBracketsGreensTheorem = SumWithinBracketsGreensTheorem
    
    #print(PositiveSumWithinBracketsGreensTheorem, "= Absolute Value of Brackets of Green's Theorem")
    
    #Please note that now that we found the inner bracket area, we have to multiply this by (1/2) to get the final area as per Green's Theorem. 
    GreensTheoremArea = (1/2)*(PositiveSumWithinBracketsGreensTheorem) 
    print("Please note the area of the polygon given by the ordered coordinates, ", coords, "is: ", GreensTheoremArea, "square units according to Green's Theorem Area! Yay! Thank you!")
    print("")
    print("")
    print("#*#*#*#*#*#*#*#*#*#*#*#*#*#*#*#*#*#*#*#*#*#*#*#*#*#*#*#*#*#*#*#*#*#*#*#*#*#*#*#*#*#*#*#*#*#*#*#*#*#*#*#*#*#*#*#*#*#*#")
    print("")
    
polygarea([(0, 0), (2,0), (2,2), (0,2)])

polygarea([(2, 1), (2,2), (3,2)])

polygarea([(2, 2), (3,2), (2,1)])
polygarea([(3, 2), (2,2), (2,1)])


polygarea([(0,0), (3, 4), (0,4), (3,0)])

polygarea([(0,0), (3, 4), (0,4), (2,6), (0,4), (3,0)])

polygarea([(0,0), (3, 4), (0,4), (2,0), (4,0), (3,0)])

polygarea([(2,1), (3, 4)])

polygarea([(2,1), (3, 1), (4, 1)])
polygarea([(2,1), (2, 4), (2, 6), (3,4)])

polygarea([(2,1), (2, 4), (2, 6), (3,4), (9, 8)])
polygarea([(0,14), (2,0), (2,3), (0,2), (9, 5), (8,7)])
```
